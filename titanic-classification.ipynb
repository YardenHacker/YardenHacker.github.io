```python
def pluralize(s):
    '''
    Modify the input string 's' to make it plural based on the following rules:
    - If the string ends with 'y', replace 'y' with 'ies' (e.g., 'city' -> 'cities').
    - If the string ends with 's', 'x', 'z', 'ch', or 'sh', add 'es' to the end.
    - Otherwise, just add 's' to the end.

    Parameters:
        s (str): The input string.

    Returns:
        str: The pluralized string.
    '''
    if s[-1]=='y':
        return s[:-1]+'ies'
    elif s[-1] in ('s','x','z'):
        return s+'es'
    elif s[-2:] in ('ch','sh') :
        return s+'es'
    else:
        return s+"s"
    
    
   
    pass
```

```python
### Use this cell to test your code
pluralize("city")
pluralize("watch")
pluralize("dash")
pluralize("dog")
```

```python
def count_substrings(s, sub):
    '''
    Count the number of times the substring 'sub' appears in the string 's', including overlapping occurrences.

    Parameters:
        s (str): The input string.
        sub (str): The substring to search for.

    Returns:
        int: The count of overlapping occurrences of 'sub'.
    '''
    if not sub:
        return 0

    count = 0
    startIndex = 0

    while True:
        i = s.find(sub, startIndex)
        if i == -1:
            return count
        count += 1
        startIndex = i + 1

    
  
```

```python
def front_back(a, b):
    '''
    Given two strings 'a' and 'b', divide each of them into two halves and then combine them
    in the following way:
    - If the length of a is even, the front and back halves are of the same length.
    - If the length of a is odd, the extra character goes in the front half.
    - The same rules apply to string 'b'.
    - Return a new string formed by concatenating a-front + b-front + a-back + b-back.

    Parameters:
        a (str): The first input string.
        b (str): The second input string.

    Returns:
        str: The combined string according to the specified rules.
    '''

   
    if(len(a)%2==0):
        front_a =a[:int(len(a)/2)]
        back_a =a[int(len(a)/2):]

    else:
        front_a =a[:int(len(a)/2+1)]
        back_a =a[int(len(a)/2+1):]

    if(len(b)%2==0):
        front_b =b[:int(len(b)/2)]
        back_b =b[int(len(b)/2):]

    else:
        front_b =b[:int(len(b)/2+1)]
        back_b =b[int(len(b)/2+1):]
    
    return(front_a+front_b+back_a+back_b)

    pass
```

```python
def reverse_and_capitalize(s):
    '''
    Given a string 's' containing multiple words separated by spaces, perform the following operations:
    - Reverse the order of the words in the string.
    - Capitalize the first letter of each word and make the rest of the letters lowercase.
    - Ensure that multiple spaces between words are reduced to a single space.
    
    Parameters:
        s (str): The input string.

    Returns:
        str: The modified string with reversed word order, adjusted capitalization, and single spaces.
    '''
    words = s.split()
    for i in range(len(words)):
        words[i] = words[i].capitalize()
    words = words[::-1]

    return " ".join(words)   
    pass
    
```

```python
def match_ends(words):
    '''
    Given a list of strings 'words', return the count of the number of
    strings where the string length is 2 or more, and the first and last
    characters of the string are the same.

    Parameters:
        words (list): A list of strings.

    Returns:
        int: The count of strings meeting the specified criteria.
    '''
    count=0
    for i in range(len(words)):
        if len(words[i])>=2:
            if words[i][0] == words[i][-1]:
                count+=1
    print(count)     
            
    pass
```

```python
def longest_common_prefix(strings):
    '''
    Given a list of strings 'strings', return the longest common prefix
    shared among all the strings. If there is no common prefix, return an empty string.

    Parameters:
        strings (list): A list of strings.

    Returns:
        str: The longest common prefix shared among all the strings.
    '''
    shortest = min(strings, key=len)
    prefix = ""
    for i in range(len(shortest)):
        ch = shortest[i]
        for s in strings:
            if s[i] != ch:
                return prefix  
        prefix += ch

    return prefix
            
                
    pass
```

```python
def remove_adjacent(nums):
    '''
    Given a list of numbers 'nums', return a list where all adjacent equal
    elements have been reduced to a single element. For example, [1, 2, 2, 3]
    returns [1, 2, 3]. You may create a new list or modify the passed-in list.

    Parameters:
        nums (list): A list of numbers.

    Returns:
        list: The modified list with adjacent equal elements removed.
    '''

    result = [nums[0]]  

    for i in range(1, len(nums)):
        if nums[i] != nums[i - 1]:   
            result.append(nums[i]) 

    return result
    pass
```

```python
def linear_merge(list1, list2):
    '''
    Merge two sorted lists into a single sorted list.

    This function takes two lists that are already sorted in increasing order
    and merges them into one list which is also sorted in increasing order.
    The function is designed to work in linear time, which means it will
    go through each list only once.

    Note: You should not use the built-in sort() or sorted() functions.

    Parameters:
        list1 (list): The first sorted list.
        list2 (list): The second sorted list.

    Returns:
        list: A merged and sorted list containing all elements from both lists.
    '''
    index1 = 0
    index2 = 0
    result = []

    total = len(list1) + len(list2)
    for i in range(total):
        if index1 >= len(list1):
            result.append(list2[index2])
            index2 += 1
            continue
        if index2 >= len(list2):
            result.append(list1[index1])
            index1 += 1
            continue
        if list1[index1] <= list2[index2]:
            result.append(list1[index1])
            index1 += 1
        else:
            result.append(list2[index2])
            index2 += 1

    return result
```
